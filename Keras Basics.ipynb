### Seqential Model

```python
import numpy as np
np.random.seed(123)
from keras.models import Sequential
from keras.layers import Dense, Activation
model = Sequential()
```

```python
#Data
x = np.random.randint(-1,2, size=(10, 3))
y = np.random.randint(2, size=(10, 1))
```

```python
# add layers
model.add(Dense(3, input_dim=3, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
```

```python
#Compile
model.compile(loss='mean_squared_error',
              optimizer='sgd',
              metrics=['accuracy'])
```

```python
# train
model.fit(x, y ,epochs=1000, steps_per_epoch=10)
```

```python
results = model.predict(x, verbose=0, steps=1)
print(results)
```

```python
test = np.array([[1,0,-1],[0,0,0]])
model.predict(test, verbose=0, steps=1)
```
